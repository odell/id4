```python
import numpy as np
import matplotlib.pyplot as plt

from finite_differences import first_order, second_order

plt.style.use('science')
```

```python
dx = 1e-3

x_min = 1e-2
x_max = 10.0
n = int((x_max - x_min) / dx) + 1
x = np.linspace(x_min, x_max, n)
```

```python
y = x*np.sin(x)
yp = np.sin(x) + x*np.cos(x)
ypp = 2*np.cos(x) - x*np.sin(x)

d1 = first_order(n, dx)
d2 = second_order(n, dx)
```

```python
abs_diff_1 = np.abs(d1 @ y - yp)
abs_diff_2 = np.abs(d2 @ y - ypp)
```

```python
ii_max = np.argsort(abs_diff_1)[::-1]
for i in ii_max[:5]:
    print(i, abs_diff_1[i])
```

```python
ii_max = np.argsort(abs_diff_2)[::-1]
for i in ii_max[:5]:
    print(i, abs_diff_2[i])
```

```python
fig, ax = plt.subplots(dpi=200)
fig.patch.set_facecolor('white')

ax.scatter(x, abs_diff_1, marker='.', alpha=0.5)
ax.scatter(x, abs_diff_2, marker='.', alpha=0.5)
ax.axhline(dx**2, color='C6')

ax.set_yscale('log');
```
