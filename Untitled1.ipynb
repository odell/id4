```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

plt.style.use('science')
```

$$
x \sim N(0, 1^2) \\
y \sim N(0, 1^2)
$$

```python
dist = stats.multivariate_normal([0, 0], np.eye(2))
```

```python
samples = dist.rvs(10000)
```

```python
fig, ax = plt.subplots(figsize=(3, 3), dpi=200)
fig.patch.set_facecolor('white')

ax.hist2d(samples[:, 0], samples[:, 1], bins=100);
```

```python
def f(x, y):
    return x + y
```

```python
z = np.array([f(*s) for s in samples])
```

```python
fig, ax = plt.subplots(figsize=(3, 3), dpi=200)
fig.patch.set_facecolor('white')

ax.hist(z, bins=50);
```
