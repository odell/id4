```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('science')
```

```python
def alpha_by_term(r, n=0):
    factor = 0
    if n >= 1:
        factor += r**5
    if n >= 2:
        factor += 9/2*r**4
    if n >= 3:
        factor += 9*r**3
    if n >= 4:
        factor += 27/2*r**2
    if n >= 5:
        factor += 27/2*r
    if n == 6:
        factor += 27/4
    return 9/2 - 2/3*np.exp(-2*r) * factor
```

```python
rs = np.linspace(0.1, 4, 1000)

fig, ax = plt.subplots(dpi=200)
fig.patch.set_facecolor('white')

for i in range(1, 7):
    ax.plot(rs, alpha_by_term(rs, i)/alpha_by_term(rs), label=rf'$n={i}$')

ax.legend()
ax.set_xlabel(r'$r$')
ax.set_ylabel(r'$V_{{\rm TL},n}/V_{\rm ID}$');
```

```python
from constants import MU, BETA4
```

```python
def beta_4(r, n=0):
    return np.sqrt(2*MU*1/2*alpha_by_term(r, n))
```

```python
tl = -np.loadtxt('spectra_Mott_Massey_hp.txt')
beta_n_tl = 1/np.sqrt(2*MU*tl)
```
